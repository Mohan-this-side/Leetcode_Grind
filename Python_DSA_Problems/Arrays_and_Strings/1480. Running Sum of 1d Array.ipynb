```python
class Solution(object):
    def runningSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        n = len(nums)
        
        prefix = [nums[0]]
        
        for i in range(1, n):
            prefix.append(nums[i] + prefix[-1])
        
        return prefix
```
