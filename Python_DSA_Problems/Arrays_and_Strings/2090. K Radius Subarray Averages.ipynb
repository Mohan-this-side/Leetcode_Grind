```python
class Solution(object):
    def getAverages(self, nums, k):
        n = len(nums)
        ans = [-1] * n
        if k < 0 or n < 2 * k + 1:
            return ans

        # Build prefix sum array without initial zero
        prefix = [nums[0]]
        for i in range(1, n):
            prefix.append(prefix[-1] + nums[i])

        for i in range(k, n - k):
            left = i - k
            right = i + k
            if left == 0:
                total = prefix[right]
            else:
                total = prefix[right] - prefix[left - 1]
            ans[i] = total // (2 * k + 1)
        return ans
```
