```python
class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        '''
        main idea is to understand the largest elements comes from ends of the array
        then adding those largest elements from backward in the ans array
        two pointer, move towards center while squaring and swapping, in the same array
        '''
        
        i = 0
        j = len(nums) - 1
        ans = [0] * len(nums)
        pos = len(nums) - 1
        
        while i <= j:
            first = (nums[i]) ** 2
            sec = (nums[j]) ** 2
            
            if first > sec:
                ans[pos] = first
                i += 1
            else:
                ans[pos] = sec
                j -= 1
                
            pos -= 1
                
        return ans
```
