```python
class Solution:
    def longestOnes(self, nums: List[int], k: int) -> int:
        curr = 0
        i = j = 0
        ans = 0
        for j in range(len(nums)):
            if nums[j] == 0:
                curr += 1
            
            while curr > k:
                if nums[i]==0:
                    curr -= 1
                    i += 1
                else:
                    i += 1
            
            ans = max(ans, j - i + 1)

        return ans
```
