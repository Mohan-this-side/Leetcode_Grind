```python
class Solution(object):
    def findMaxAverage(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: float
        """
        curr = 0
        
        for right in range(k):
            curr += nums[right]
            
            
        ans = curr  

        for right in range(k, len(nums)):
            curr += nums[right] - nums[right - k]
            ans = max(ans, curr)
            
        return ans / float(k)
```
